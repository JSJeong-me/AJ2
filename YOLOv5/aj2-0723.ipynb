```python
import torch
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
img = 'elephant2.png'
```

```python
result = model(img)
```

```python
df = result.pandas().xyxy[0]
```

```python
df
```

```python
x1 = int(df.iloc[0].xmin)
x2 = int(df.iloc[0].xmax)
y1 = int(df.iloc[0].ymin)
y2 = int(df.iloc[0].ymax)
```

```python
ele1 = src[y1:y2, x1:x2]
```

```python
cv2_imshow(ele1)
```
